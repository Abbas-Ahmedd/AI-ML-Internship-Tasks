# Task 1 Exploring and Visualizing Iris Dataset

## Problem Statement
The goal of this task is to explore and visualize the Iris dataset to understand feature distributions, relationships between variables, and detect possible outliers.

## Objective
To use pandas, matplotlib, and seaborn for data loading, inspection, descriptive statistics, and visualization.|

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Iris dataset from seaborn
df = sns.load_dataset("iris")

# Display first 5 rows
df.head()
```

```python
print("Missing Values:")
print(df.isnull().sum())
```

```python
print("Shape of dataset:", df.shape)
```

```python
print("Column Names:")
print(df.columns)
```

```python
df.info()
```

```python
df.describe()
```

## Scatter Plot

This plot helps visualize the relationship between sepal length and petal length across different species.

```python
plt.figure(figsize=(8,6))
sns.scatterplot(
    x='sepal_length',
    y='petal_length',
    hue='species',
    data=df
)

plt.title("Scatter Plot of Sepal Length vs Petal Length")
plt.show()
```

## Histogram

Histograms help show the distribution of values for each feature in the Iris dataset, such as sepal length, petal length, and petal width.

```python
df.hist(figsize=(10,8))
plt.suptitle("Histograms of Iris Features")
plt.show()
```

## Box Plot

Box plots help identify outliers and show the spread of values across different numerical features.

```python
plt.figure(figsize=(8,6))
sns.boxplot(data=df)

plt.title("Box Plot for Detecting Outliers")
plt.show()
```
